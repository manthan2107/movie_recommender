```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Read data
data = pd.read_csv('run_time.csv', header=None, names=['Method', 'Time (seconds)'])

# Plotting each method attempt vs time
fig, ax = plt.subplots(figsize=(10, 6))
for method, group in data.groupby('Method'):
    attempts = np.arange(1, len(group) + 1)
    ax.plot(attempts, group['Time (seconds)'], marker='o', label=method)

ax.set_xlabel('Attempt')
ax.set_ylabel('Time (seconds)')
ax.set_title('Method Attempt vs Time')
ax.legend()
plt.grid(True)
plt.show()

# Calculate statistics
statistics = data.groupby('Method')['Time (seconds)'].agg(['min', 'max', 'mean', 'median', 'std'])
print(statistics)
```
